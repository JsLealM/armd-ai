# Organizacion de CSV procesados

## Objetivo

Este notebook no solo documenta la organizacion de `3.DATOS-PROCESADOS`.
Tambien permite ejecutarla y verificarla dentro de Jupyter Lab.

Regla importante del proyecto:

- los resultados deben quedar explicados en notebook
- si se usa un script `.py`, el notebook debe mostrar como ejecutarlo y que salida produjo


```python
from pathlib import Path
import subprocess
import sys

RUTA_PROYECTO = Path.cwd().resolve().parent
RUTA_PROCESADOS = RUTA_PROYECTO / "3.DATOS-PROCESADOS"
RUTA_SCRIPT = RUTA_PROCESADOS / "organizar_csv_procesados.py"
RUTA_INDICE = RUTA_PROCESADOS / "INDICE_CSV_PROCESADOS.md"

print(RUTA_PROCESADOS)
print(RUTA_SCRIPT)
```

## Ejecutar la organizacion

La siguiente celda ejecuta el script real que organiza los CSV en subcarpetas sin mover ni romper los archivos originales.

Importante:

- los nombres internos de la raiz de `3.DATOS-PROCESADOS` se conservan para no romper notebooks ni scripts
- las copias organizadas en subcarpetas si usan nombres numerados y mas faciles de leer


```python
resultado = subprocess.run(
    [sys.executable, str(RUTA_SCRIPT)],
    cwd=RUTA_PROCESADOS,
    capture_output=True,
    text=True,
    check=True,
)
print(resultado.stdout)
if resultado.stderr.strip():
    print('STDERR:')
    print(resultado.stderr)
```

## Verificar subcarpetas y CSV copiados


```python
for carpeta in sorted([p for p in RUTA_PROCESADOS.iterdir() if p.is_dir() and not p.name.startswith('.')]):
    print(f"\n[{carpeta.name}]")
    for archivo in sorted(carpeta.glob('*.csv')):
        print('-', archivo.name)
```

## Leer el indice generado

Esto permite ver en notebook el resumen de la organizacion.


```python
print(RUTA_INDICE.read_text(encoding='utf-8'))
```
