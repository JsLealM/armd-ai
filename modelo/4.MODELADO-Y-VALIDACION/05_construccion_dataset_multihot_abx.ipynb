# Construccion de dataset multihot de antibioticos previos

## Objetivo

Este notebook deja documentada y ejecutable la construccion de la variante `multihot_abx`.

La regla del proyecto se mantiene: el proceso importante no debe depender de que alguien corra un `.py` por fuera de Jupyter Lab.


## Diseño metodologico

La exposicion antibiotica previa no se representa aqui como una sola columna `antibiotic_class`.

En su lugar, se genera una codificacion `multihot` por clase antibiotica previa detectada en la cohorte.

Importante:

- no se usa recencia temporal
- `time_to_culturetime` no entra en esta variante final


```python
from pathlib import Path
import subprocess
import sys
import pandas as pd

RUTA_PROYECTO = Path.cwd().resolve().parent
RUTA_MODELADO = RUTA_PROYECTO / "4.MODELADO-Y-VALIDACION"
RUTA_PROCESADOS = RUTA_PROYECTO / "3.DATOS-PROCESADOS"
RUTA_SCRIPT = RUTA_MODELADO / "construir_dataset_multihot_abx.py"

RUTA_DATASET = RUTA_PROCESADOS / "armd_s_aureus_base_modelado_multihot_abx.csv"
RUTA_MANIFIESTO = RUTA_PROCESADOS / "manifiesto_modelado_multihot_abx.csv"
RUTA_CATALOGO = RUTA_PROCESADOS / "catalogo_clases_antibioticas_multihot.csv"

print(RUTA_SCRIPT)
```

## Ejecutar construccion

La siguiente celda ejecuta el proceso real y genera los archivos finales de esta variante.


```python
resultado = subprocess.run(
    [sys.executable, str(RUTA_SCRIPT)],
    cwd=RUTA_MODELADO,
    capture_output=True,
    text=True,
    check=True,
)
print(resultado.stdout)
if resultado.stderr.strip():
    print('STDERR:')
    print(resultado.stderr)
```

## Verificar artefactos generados

Aqui se revisa el dataset final, el manifiesto y el catalogo de clases antibioticas.


```python
df = pd.read_csv(RUTA_DATASET, low_memory=False)
print('Shape:', df.shape)
columnas_abx = [c for c in df.columns if c.startswith('exp_prev_')]
print('Columnas multihot:', len(columnas_abx))
print(columnas_abx)
```
